Standardization

```python
import numpy as np
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df = df.iloc[:, 2: ]
```

```python
df.sample(5)
```

```python
df.head()
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.drop('Purchased', axis = 1),
                                                    df['Purchased'],
                                                    test_size = 0.2,
                                                    random_state = 0 )
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
X_train.head(5)
```

```python
Y_train.head()
```

```python
X_test.head(5)
```

StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit the scaler to the training data
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train_scaled
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test.columns)
```

```python
X_train_scaled.head(5)
```

Effect of Scaling

```python
import matplotlib.pyplot as plt
fig,(ax1,ax2) = plt.subplots(ncols = 2, figsize = (10,6))
ax1.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'])
ax1.set_title('after Scaling')
ax2.scatter(X_train['Age'], X_train['EstimatedSalary'],color = 'orange')
ax2.set_title('before Scaling')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(ncols = 2, figsize = (10,6))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x = X_train['Age'], ax = ax1)
sns.kdeplot(x = X_train['EstimatedSalary'], ax = ax1)

# after scaling
ax2.set_title('After Scaling')
sns.kdeplot(x = X_train_scaled['Age'], ax = ax2)
sns.kdeplot(x = X_train_scaled['EstimatedSalary'], ax = ax2)

plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols = 2, figsize = (10,6))
sns.kdeplot(x = X_train['Age'], ax = ax1)
sns.kdeplot(x = X_train_scaled['Age'], ax = ax2)
ax1.set_title('before Scaling')
ax2.set_title('after Scaling') 
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 6))
#before scaling
sns.kdeplot(x = X_train['EstimatedSalary'], ax = ax1)
ax1.set_title('before Scaling')
#after scaling
sns.kdeplot(x = X_train_scaled['EstimatedSalary'], ax = ax2)
ax2.set_title('after Scaling') 
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, Y_train)
lr_scaled.fit(X_train_scaled, Y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
print("actual",accuracy_score(Y_test, y_pred))
print("scaled",accuracy_score(Y_test, y_pred_scaled))
```
